```python
import requests as rq

record = []

for i in range(1,6):
    url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={i}&sparkline=False"
    response = rq.get(url)

    if response.status_code == 200:
        data = response.json()
        record.extend(data)    
```

```python
coin_id = ["bitcoin","ethereum","tether"]
rows = []
pd.options.display.float_format = '{:.2f}'.format

for i in coin_id:
    url = f"https://api.coingecko.com/api/v3/coins/{i}/market_chart?vs_currency=inr&days=365"
    response = rq.get(url)

    if response.status_code == 200:
        data = response.json()
    
        for timestamp, price in data['prices']:
            rows.append({
                'coin_id': i,
                'timestamp': timestamp,
                'price_inr': price
            })
        else:
            print('Limit Exceeded')

top_3_coins = pd.DataFrame(rows)
top_3_coins['timestamp'] = pd.to_datetime(top_3_coins['timestamp'], unit='ms').dt.date
top_3_coins
```

```python
# 1. Find the top 3 cryptocurrencies by market cap.

query = 'SELECT * FROM cryptocurrencies ORDER BY market_cap DESC LIMIT 3'
df_top3 = pd.read_sql(query, conn)
print(df_top3)
```

```python
# 2. List all coins where circulating supply exceeds 90% of total supply.

query = 'SELECT id FROM cryptocurrencies WHERE circulating_supply > (total_supply * 0.90)'
df_topcoins = pd.read_sql(query, conn)
print(df_topcoins)
```

```python
# 3. Get coins that are within 10% of their all-time-high (ATH)

query = 'select * from cryptocurrencies where ath <= (select max(ath)*0.10 from cryptocurrencies)'
df_max_ath = pd.read_sql(query, conn)
print(df_max_ath)
```

```python
# 4. Find the average market cap rank of coins with volume above $1B.

query = 'SELECT AVG(market_cap_rank) AS average_rank FROM cryptocurrencies WHERE total_volume > 1000000000'
df_avg_market_cap = pd.read_sql(query, conn)
print(df_avg_market_cap)
```

```python
# 5. Get the most recently updated coin.

query = 'SELECT id, date FROM cryptocurrencies order by date desc limit 1'
df_recent_update_coin = pd.read_sql(query, conn)
print(df_recent_update_coin)
```

```python
# 1. Find the highest daily price of Bitcoin in the last 365 days.

query = "select timestamp, max(price_inr) max_price from crypto_prices where coin_id = 'bitcoin' group by timestamp order by timestamp desc"
df_recent_update_coin = pd.read_sql(query, conn)
print(df_recent_update_coin)
```

```python
#2.Calculate the average daily price of Ethereum in the past 1 year

query = "select timestamp, avg(price_inr) avg_price from crypto_prices where coin_id = 'Ethereum' group by timestamp order by timestamp desc"
df_avg_price_coin = pd.read_sql(query, conn)
print(df_avg_price_coin)
```

```python
#3. Show the daily price trend of Bitcoin in January 2026

query = "select * from crypto_prices where coin_id = 'bitcoin' and timestamp between '2026-01-01' and '2026-01-31' order by timestamp"
df_jan_price_coin = pd.read_sql(query, conn)
print(df_jan_price_coin)
```

```python
#4. Find the coin with the highest average price over 1 year.
query = "select coin_id, avg(price_inr) from crypto_prices group by coin_id order by 2 desc limit 1"
df_high_avg_coin = pd.read_sql(query, conn)
print(df_high_avg_coin)
```

```python
#5. Get the % change in Bitcoin’s price between May 2025 and Mar 2026

query = '''WITH StartPrice AS (
    SELECT price_inr 
    FROM crypto_prices 
    WHERE coin_id = 'bitcoin' AND timestamp = '2025-05-01'
),
EndPrice AS (
    SELECT price_inr 
    FROM crypto_prices 
    WHERE coin_id = 'bitcoin' AND timestamp = '2026-03-31'
)
SELECT 
    ((e.price_inr - s.price_inr) / s.price_inr) * 100 AS percentage_change
FROM StartPrice s, EndPrice e'''
df_change_per_coin = pd.read_sql(query, conn)
print(df_change_per_coin)
```

```python
#1. Find the highest oil price in the last 5 years.

    query = '''SELECT MAX(price) AS highest_price
    FROM oil_prices
    WHERE date BETWEEN (SELECT DATE_SUB(MAX(date), INTERVAL 5 YEAR) FROM oil_prices) 
                   AND (SELECT MAX(date) FROM oil_prices)'''
df_high_oil_price = pd.read_sql(query, conn)
print(df_high_oil_price)
```

```python
#2. Get the average oil price per year.

query = '''select year(date), avg(price) avg_oil_price  from oil_prices group by year(date) order by 1 desc'''
df_avg_oil_price = pd.read_sql(query, conn)
print(df_avg_oil_price)
```

```python
#3. Show oil prices during COVID crash (March–April 2020)

query = '''select * from oil_prices where date >= '2020-03-01' and date <= '2020-04-30' order by date'''
df_covid_crash_price = pd.read_sql(query, conn)
print(df_covid_crash_price)
```

```python
#4. Find the lowest price of oil in the last 10 years.

query = '''SELECT MIN(price) AS highest_price
FROM oil_prices
WHERE date BETWEEN (SELECT Min(date) FROM oil_prices) 
               AND (SELECT MAX(date) FROM oil_prices)'''
df_low_oil_price = pd.read_sql(query, conn)
print(df_low_oil_price)
```

```python
#5. Calculate the volatility of oil prices (max-min difference per year).

query = '''select year(date), max(price) - min(price) price_volatility  from oil_prices group by year(date) order by 1 desc'''
df_voilatile_price = pd.read_sql(query, conn)
print(df_voilatile_price)
```

```python
# Join Queries (Cross-Market Analysis)
```

```python
#1. Compare Bitcoin vs Oil average price in 2025.

query =  '''select '2025' year
		, (select avg(price_inr) from crypto_prices where coin_id = 'bitcoin' and year(timestamp) = '2025') bitocoin_avg_price
		, (select avg(price) from oil_prices where year(date) = '2025') oil_avg_price '''
df_avg_price = pd.read_sql(query, conn)
print(df_avg_price)
```

```python
#2. Check if Bitcoin moves with S&P 500 (correlation idea).

query =  '''select s.ticker, cast(s.date as date) date, s.open sp500_price, c.coin_id, c.price_inr bitcoint_price
from stock_prices s inner join crypto_prices c on s.date = c.timestamp
where ticker = '^GSPC' and coin_id = 'bitcoin'
order by s.date desc'''
df_bitcoin_sp500 = pd.read_sql(query, conn)
print(df_bitcoin_sp500)
```

```python
#3. Compare Ethereum and NASDAQ daily prices for 2025.

query =  '''SELECT 
    c.coin_id, 
    c.timestamp, 
    c.price_inr ethereum_price, 
    s.ticker, 
    s.close nasdaq_price
FROM crypto_prices c
INNER JOIN stock_prices s 
    ON DATE(c.timestamp) = s.date
WHERE c.coin_id = 'Ethereum' 
  AND s.ticker = '^IXIC'
  AND YEAR(s.date) = 2025
ORDER BY s.date'''
df_daily_price = pd.read_sql(query, conn)
print(df_daily_price)
```

```python
#4. Find days when oil price spiked and compare with Bitcoin price change

query =  '''select c.coin_id, c.timestamp, c.price_inr bitcoin_price, o.price oil_price
from crypto_prices c inner join oil_prices o on (c.timestamp = o.date and o.price > c.price_inr)
where coin_id = 'bitcoin' order by c.timestamp desc'''
df_oil_bitocoin = pd.read_sql(query, conn)
print(df_oil_bitocoin)

```

```python
#5.Compare top 3 coins daily price trend vs Nifty (^NSEI).

query =  '''select s.ticker, cast(s.date as date) date, s.open, c.coin_id, c.price_inr
from stock_prices s inner join crypto_prices c on s.date = c.timestamp
where ticker = '^NSEI'
order by s.date desc'''
df_topcoins_nifty = pd.read_sql(query, conn)
print(df_topcoins_nifty)
```

```python
#6.Compare stock prices (^GSPC) with crude oil prices on the same dates

query =  '''select s.ticker, cast(s.date as date) date, s.open ticker_price,  o.price oil_price
from stock_prices s inner join oil_prices o on s.date = o.date
where ticker = '^GSPC'
order by s.date desc'''
df_crudeoil_gspc = pd.read_sql(query, conn)
print(df_crudeoil_gspc)
```

```python
#7.Correlate Bitcoin closing price with crude oil closing price (same date)

query =  '''select c.coin_id, c.timestamp, c.price_inr bitcoin_price, o.price oil_price
from crypto_prices c inner join oil_prices o on c.timestamp = o.date
where coin_id = 'bitcoin' order by c.timestamp desc'''
df_bitcoin_oil_close = pd.read_sql(query, conn)
print(df_bitcoin_oil_close)
```

```python
#8.Compare NASDAQ (^IXIC) with Ethereum price trends

query =  '''SELECT c.coin_id, c.timestamp date, c.price_inr ethereum_price, s.ticker, s.close nasdaq_price
FROM crypto_prices c INNER JOIN stock_prices s ON DATE(c.timestamp) = s.date
WHERE c.coin_id = 'Ethereum' AND s.ticker = '^IXIC'
ORDER BY s.date desc'''
df_ixic_eretherum_price = pd.read_sql(query, conn)
print(df_ixic_eretherum_price)
```

```python
#9.Join top 3 crypto coins with stock indices for 2025

query =  '''SELECT c.coin_id, c.timestamp date, c.price_inr crypto_price, s.ticker, s.close stock_price
FROM crypto_prices c INNER JOIN stock_prices s ON DATE(c.timestamp) = s.date
ORDER BY s.date desc'''
df_coins_stock_price = pd.read_sql(query, conn)
print(df_coins_stock_price)
```

```python
#10. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison

query =  '''select c.coin_id, c.timestamp, c.price_inr bitcoin_price, o.price oil_price, s.ticker, s.close stock_price
from crypto_prices c inner join oil_prices o on c.timestamp = o.date
	inner join stock_prices s on o.date = s.date
where coin_id = 'bitcoin' order by c.timestamp desc'''
df_bitcoin_stock_oil_price = pd.read_sql(query, conn)
print(df_bitcoin_stock_oil_price)
```
